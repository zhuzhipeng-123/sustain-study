```python
import numpy as np
import random
matrix = np.random.rand(5, 5)

n=20#迭代轮数

#该算法类似GD，会慢慢找到最优解，通过迭代的方式

for i in range(n):
    matrix=matrix/matrix.sum(axis=1,keepdims=True)
    matrix=matrix/matrix.sum(axis=0,keepdims=True)
    a1=max(matrix.sum(axis=1,keepdims=True))
    a2=max(matrix.sum(axis=0,keepdims=True))



    print('最小的',a1,'最大的数',a2)
```
